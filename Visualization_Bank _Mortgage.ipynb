```python
# Import file

import pandas as pd

df = pd.read_csv('sample_bankmortgage_data.csv')

df.head(10)
```

```python
# Convert YES/NO values into 0 and 1

df['pep'] = df['pep'].map({'NO': 0, 'YES': 1})
df['car'] = df['car'].map({'NO': 0, 'YES': 1})

df.head(10)
```

```python
# Perform min-max calculation on the income column 

df['income'] = (df['income'] - df['income'].min()) / (df['income'].max() - df['income'].min())

df
```

```python
# Group the rows by Gender and calculate the mean of selected columns for each group 

group_by_gender = df.groupby('sex')[['income','car','pep']].mean()

group_by_gender
```

```python
# Visualization of Gender vs Equity Plan(based on income, car and pep)

import matplotlib.pyplot as plt

group_by_gender.plot.bar(title='Gender vs Equity Plan')

plt.ylabel('Equity Plan')
plt.xlabel('Gender')

plt.show()
```

```python
# Convert YES/NO values into 0 and 1

df['mortgage'] = df['mortgage'].map({'NO': 0, 'YES': 1})

df
```

```python
# Create a pivot table to summarize the data from each region 

piv = pd.pivot_table(df, values='mortgage', index=['region'])

piv_table = piv.rename(columns={'mortgage':'Mortgage Liability'})

piv_table
```

```python
# Generate a heatmap from the values in the pivot table 

import seaborn as sns

sns.heatmap(piv_table, cmap="YlGnBu", square=True)

plt.title('Correlation for Region and Mortgage Liability')
plt.ylabel('Region')

plt.show()
```

```python
# Filter specific values from the selected columns

new_filter = df[(df['age'] >= 20) & (df['age'] <= 30) & (df['mortgage'] == 0) & (df['children'] == 0)]

new_filter.head(10)
```

```python
# Create a node and edge graph 

import networkx as nx

G = nx.Graph()

for index, row in new_filter.iterrows():
  age = row['age']
  G.add_node(age)
  G.add_edge(age,'No mortgage and no children')

nx.draw(G, with_labels=True)
plt.show()
```
